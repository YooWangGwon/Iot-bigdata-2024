##### matplotlib은 기본적으로 한글 깨짐
- 한글 폰트 연결 필요

- 폰트 관리자 설정

````python
#Matplotlib 사용
import matplotlib.pyplot as plt # plt로 줄여씀
from matplotlib import rcParams, font_manager, rc
import warnings

# 나눔폰트 경로
font_path = 'C:/Windows/Fonts/malgun.ttf'
font = font_manager.FontProperties(fname = font_path, size=10).get_name()

# warning 표시 끄기
warnings.simplefilter('ignore') # 경고 메시지 무시

# 기본 폰트 설정
plt.rcParams['font.family'] = font

# 기본 폰트 사이즈 설정
plt.rcParams['font.size'] = 14

# 기본 차트 사이즈 변경
plt.rcParams['figure.figsize'] = (8, 6)

# 기본 그리드 표시
plt.rcParams['axes.grid'] = True

# 한글화 후 마이너스 표시 네모를 정상출력
plt.rcParams['axes.unicode_minus'] = False
````

````python
import numpy as np # 보통 np로 줄여씀
import seaborn as sns

# style => dark, darkgrid, white, whitegrid, ticks
sns.set(font=font, rc ={'axes.unicode_minus':False}, style="darkgrid")

# 스타일 한번 지정하면 그 뒤 모든 차트는 시본 차트가 됨
````

````python
## 스캐터 차트
x = np.random.randint(-100, 100, 50) # -100 ~ 100 사이에 50개만 랜덤
y = np.random.randint(-200, 200, 50) # -200 ~ 200 사이에 50개 랜덤
size = np.random.randint(-200, 200, 50)
color = np.random.randint(0, 100, 50)

plt.scatter(x, y, s=size, c=color, alpha=0.5)
plt.title('온도(화씨)')
plt.colorbar()
plt.show()
````

````python
## 파이차트
size = [2441, 2312, 1031, 1233]
label = ['A', 'B', 'AB', 'O']
color = ['darkmagenta', 'deeppink', 'hotpink', 'pink']

plt.axis('equal') # 가로축, 세로축 스케일 동일
plt.pie(size, labels=label, autopct='%.1f%%', explode=(0,0,0.1,0), colors=color)
plt.legend(loc='lower left')
plt.title('혈액형별 인구율')
plt.show()
````
